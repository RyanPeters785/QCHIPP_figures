```python
import matplotlib.pyplot as plt
```

```python
# number of amino acids encoded
aas = [1,2,3,4,5,6,7,8,9]

# number of qubits required using each (one-hot/binary encoding with basis/qrac2:1 feature map)
oneh_basis = [20,40,60,80,100,120,140,160,180]
oneh_qrac = [10,20,30,40,50,60,70,80,90]
binary_basis = [5,10,15,20,25,30,35,40,45]
binary_qrac = [2.5,5,7.5,10,12.5,15,17.5,20,22.5]

# round to proper number of qubtis: 
# - the qrac 2:1 feature map requires an even number of classical bits to map properly to qubits
# - the qunatum circuit requires an even number of qubits after being mapped to quantum states via feature map
for i in range(len(oneh_basis)):
    while oneh_basis[i] % 2 != 0:
        oneh_basis[i] += 1
for i in range(len(oneh_qrac)):
    while oneh_qrac[i] % 4 != 0:
        oneh_qrac[i] += 1
for i in range(len(binary_basis)):
    while binary_basis[i] % 2 != 0:
        binary_basis[i] += 1
for i in range(len(binary_qrac)):
    while (binary_qrac[i] * 10) % 40 != 0:
        binary_qrac[i] += 0.5
        

# colors
oneh_basis_color = (131/255, 151/255, 92/255)
oneh_qrac_color = (98/255, 131/255, 155/255)
binary_basis_color = (191/255, 61/255, 36/255)
binary_qrac_color = (236/255, 183/255, 3/255)


# create figure
plt.figure(figsize=(8, 6))

plt.plot(aas, oneh_basis, color=oneh_basis_color, marker='o', label='One-Hot + Basis', linewidth=2, markersize=6)
plt.plot(aas, oneh_qrac, color=oneh_qrac_color, marker='o', label='One-Hot + QRAC 2:1', linewidth=2, markersize=6)
plt.plot(aas, binary_basis, color=binary_basis_color, marker='o', label='Binary + Basis', linewidth=2, markersize=6)
plt.plot(aas, binary_qrac, color=binary_qrac_color, marker='o', label='Binary + QRAC 2:1', linewidth=2, markersize=6)

plt.xlabel('# of Amino Acids', fontsize=18)
plt.ylabel('Qubits Required', fontsize=18)

plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(loc='best', fontsize=14)
plt.tight_layout()

plt.savefig(f'c_enocding_schemes.svg', dpi=300, bbox_inches='tight')
plt.show()
```
