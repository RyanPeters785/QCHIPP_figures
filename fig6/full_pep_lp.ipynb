```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load loss values
loss_values = np.load('../../qcnn/experiments/exp10_immuno_all_pos/qc_all_positions_basis_20k_objFuncVal.npy')
iterations = list(range(len(loss_values)))

# grab every other loss value (there are 2 loss values for each iteration of SPSA --> use the last as the loss)
loss_values = loss_values[1::2]
iterations = [i for i in range(len(loss_values))]

# set font and style
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16

# colors
sim_color = (131/255, 151/255, 92/255)
hardware_color = (98/255, 131/255, 155/255)

# plot loss vlaues
plt.figure(figsize=(8, 6))
plt.scatter(iterations, loss_values, color=hardware_color, label='Quantum Hardware', s=30)

plt.xlabel('Iteration', fontsize=18)
plt.ylabel('Loss', fontsize=18)

plt.ylim(-6, 1)  
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(loc='best', fontsize=14)
plt.tight_layout()

plt.savefig(f'full_pep_lp.svg', dpi=300, bbox_inches='tight')
plt.show()
```
